```python
%matplotlib inline

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.style.use('ggplot')
```

```python
with open('../src1/pickle_files/features.pkl','rb') as f:
    features = pickle.load(f)
```

```python
with open('../src1/pickle_files/labels.pkl', 'rb') as l:
    labels = pickle.load(l)
```

```python
features.shape
```

```python
pca = PCA(n_components=2)
```

```python
features.shape
```

```python
transfer_values_reduced = pca.fit_transform(features)
```

```python
transfer_values_reduced.shape
```

```python
def scatter_helper(values,cls):
    color_map = cm.rainbow(np.linspace(0.0,1.0,4))
    colors = color_map[cls]
    x = values [:, 0]
    y = values[:, 1]
    
    plt.scatter(x, y, color=colors)
    plt.legend(cls)
    plt.show()
```

```python
scatter_helper(transfer_values_reduced,cls)
```

```python
cls = []
for element in labels:
    if element == 'kickflip':
        cls.append(1)
    else:
        cls.append(0)
```
